```python
from transquest.models.sentence_model import SentenceModel

# Load the English-Hindi QE model
model = SentenceModel("TransQuest/monotransquest-da-en_hi", use_cuda=False)

# Example sentences
source_sentences = ["The cat is sleeping."]
translated_sentences = ["बिल्ली सो रही है।"]

# Predict QE score
scores = model.predict(list(zip(source_sentences, translated_sentences)))

print("\n🔹 **Quality Estimation Score:**", scores[0])

```

```python
from comet import download_model, load_from_checkpoint

# Load COMET-QE model
model_path = download_model("Unbabel/wmt20-comet-qe-da")
model = load_from_checkpoint(model_path)

# Example sentence
source = ["The dog is sleeping."]
hypothesis = ["कुत्ता सो रहा है।"]

# Compute QE score
data = [{"src": src, "mt": mt} for src, mt in zip(source, hypothesis)]
scores = model.predict(data)

print("\n🔹 **Quality Estimation Score:**", scores)
```
