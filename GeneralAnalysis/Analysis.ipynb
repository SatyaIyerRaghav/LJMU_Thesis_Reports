```python
# Merge train and dev data on the index (or sentence ID if available)
merged_df = pd.merge(
    enhitrain[['index', 'mean']],
    enhidev[['index', 'mean']],
    on='index',
    suffixes=('_train', '_dev')
)
# Scatter plot
plt.figure(figsize=(4,4))
sns.scatterplot(data=merged_df, x='mean_train', y='mean_dev')
plt.title('Merged QE Score Correlation: Train vs Dev (Mean)[En-Hi]')
plt.xlabel('Train Mean')
plt.ylabel('Dev Mean')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Use a safe fallback if no common 'index' column:
merged_df = pd.concat([
    enmrtrain[['mean']].reset_index(drop=True).rename(columns={'mean': 'mean_train'}),
    enmrdev[['mean']].reset_index(drop=True).rename(columns={'mean': 'mean_dev'})
], axis=1)
```

```python
# Scatter plot
plt.figure(figsize=(4,4))
sns.scatterplot(data=merged_df, x='mean_train', y='mean_dev')
plt.title('Merged QE Score Correlation: Train vs Dev (Mean)[En-Mr]')
plt.xlabel('Train Mean')
plt.ylabel('Dev Mean')
plt.grid(True)
plt.tight_layout()
plt.show()
```
