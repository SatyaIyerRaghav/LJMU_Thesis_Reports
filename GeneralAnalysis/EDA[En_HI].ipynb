```python
####### TRANSQUEST[EN-HI] #######
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trans=pd.read_csv('TransQE[en-hi].csv',encoding='latin1')
```

```python
trans.columns
```

```python
# Predicted Score vs Human Score Plot (Dev Set)
```

```python
#1).TRANSQUEST
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(data=trans, x='mean', y='QE_score')
plt.title('Model Predictions vs Human Scores')
plt.xlabel('Transquest Human QE Score (mean)[En-Hi]')
plt.ylabel('Transquest Model Predicted Score[En-Hi]')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Error Distribution Histogram (Optional)
```

```python
trans['abs_error'] = abs(trans['mean'] - trans['QE_score'])

plt.figure(figsize=(6, 4))
sns.histplot(trans['abs_error'], bins=30, kde=True)
plt.title('Transquest Absolute Error Distribution[En-Hi]')
plt.xlabel('Transquest Absolute Error (|predicted - actual|)[en-hi]')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap (Optional)
```

```python
# Compute correlations
corr_df = trans[['mean', 'QE_score']].corr()

plt.figure(figsize=(4, 3))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Transquest Correlation: Human vs Predicted Scores [En-Hi]')
plt.tight_layout()
plt.show()
```
